## __Abstract__

### Background

_**What has to be stated:**_

1. Studies happen

3. In the FDA publication they stated that data inconclusive and insignificant. Independent analyses showed that as well:

So it's been a concern.

> In clinical studies, adverse reactions in participants 16 years of age and older included pain at the injection site (84.1%), fatigue (62.9%), headache (55.1%), muscle pain (38.3%), chills (31.9%), joint pain (23.6%), fever (14.2%), injection site swelling (10.5%), injection site redness (9.5%), nausea (1.1%), malaise (0.5%), and lymphadenopathy (0.3%).

> Severe allergic reactions, including anaphylaxis, have been reported following the Pfizer-BioNTech COVID-19 Vaccine during mass vaccination outside of clinical trials.


### Objective

1. Data is indeed inconclusive if the sample is not biased. But what if it is biased?

2. The objective is: to figure out possible bias in the study, and make statistical analysis based on that to conclude whether the case of incidence of BP in the study is significant, and what the effect is.

### Methods

Binomial tests are used to test the hypothesis assess the significance of effect on incidence and prevalence. Extremely small probablities were hypothesised, so Z-score approximation for binomial tests was not used.

$\alpha$-level of 1% is chosen for assesing significance in difference in incidence and prevalence, because there are plenty of other diseases out there that could get spiked for this study.

$\alpha$-level of 5% is chosen for difference between proportions.



### Results

By hypothesizing bias in the Phase 3 trials of Pfizer-BioNTech COVID-19 vaccine based on their data, I challenged the significance of incidence of Bell's palsy in the vaccine group

Immediately two the most important notes here:

1) «In the United States, the annual incidence of Bell's palsy is approximately 23 cases per 100,000 persons»

23/100000 annually = 0.00023 annually = 0.023% annually

(from doi): «a total of 37,706 participants had a median of at least of 2 month of screening». At the same time, people've been screened «between July 27, 2020, and November 14, 2020», which is about up to 3.5 month.

Bell's palsy is relatively concerning condition, so it's logical to assume that people are more likely to report it, and that the average duration of screening longer than the overall average duration of screening. Given a _median_ of at least of 2 months of screening, we can safely assume an _average_ of 2 months of availability to report a case of Bell's palsy.

2) «Very few cases are observed during the summer months»

From overall of 3.5 months of screening 1+ was in summer and 2+ were in fall. According to some data on seasonal effects on incidence of Bell's palsy:

 - https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3470850/ (data from: US, Italy, Greece)
 - https://www.karger.com/Article/FullText/492097 (data from: Germany)

, during summer there are significantly less BP than on average, and during fall it's about average. Therefore the expected number of cases of Bell's palsy during the study should be lower than on average. Assuming average incidence (23 of 100'000 per year), we likely round the number towards value of _less effect_.
